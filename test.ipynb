```python
import torch
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = "checkpoints/Wav-CBT_base_abl1_GF2/Wav-CBT_base_abl1_GF2_2024_06_03-00_36_10.pth.tar"
```

```python
checkpoint = torch.load(path)
```

```python
for key, value in checkpoint.items():
        print(f"{key}: {type(value)}")
```

```python
checkpoint['val_metrics']
```

```python
val_loss = [l["loss"] for l in checkpoint['val_metrics']]
val_psnr = [l["psnr"] for l in checkpoint['val_metrics']]
val_ssim = [l["ssim"] for l in checkpoint['val_metrics']]
```

```python
plt.plot(val_loss)
plt.title("val_loss")
plt.show()
```

```python
plt.plot(val_ssim)
plt.title("val_ssim")
plt.show()
```

```python
plt.plot(val_psnr)
plt.title("val_psnr")
plt.show()
```
